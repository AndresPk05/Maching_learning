```python
import numpy as np
import matplotlib.pyplot as mtl
```

```python
data = np.random.random(size=(30,2))
print(data)
```

```python
mtl.scatter(data[:,0], data[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
def kmeans(data, k):
  kmeans = KMeans(n_clusters=4).fit(data);
  return kmeans;
def centroids(kmeans):
  centroids = kmeans.cluster_centers_;
  return centroids;
```

```python
def renderGraf(data, centroids, kmeans):
  mtl.scatter(data[:,0], data[:,1], c=kmeans.labels_.astype(float), s = 50)
  mtl.scatter(centroids[:,0], centroids[:,1], c = 'red', marker = '*', s = 50)
  mtl.show()
```

```python
k = input('digite la cantidad de clusters: ')
```

```python
kmeans = kmeans(data, k)
centroids = centroids(kmeans)
print(centroids)
```

```python
renderGraf(data, centroids, kmeans)
```

```python
clust = kmeans.predict(data);
print(clust);
```

```python
krang = range(1, 10);
sse = []
for n in krang:
  kmeans = KMeans(n_clusters=n).fit(data);
  sse.append(kmeans.inertia_);
```

```python
sse
```

```python
mtl.xlabel('K');
mtl.ylabel('SSE');
mtl.plot(krang, sse);
```
